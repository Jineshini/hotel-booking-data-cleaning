```python
# Install and import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', 35)

# Load the dataset
df = pd.read_csv('/content/hotel_bookings.csv')  # Adjust path if necessary
```

```python
# Shape of the dataset
print("Shape of dataset:", df.shape)

# Data types of each column
print("\nData types:\n", df.dtypes)

# First 5 rows
print("\nFirst 5 rows:")
display(df.head())

# Last 5 rows
print("\nLast 5 rows:")
display(df.tail())

# Basic statistical summary
print("\nStatistical Summary:")
display(df.describe(include='all'))
```

```python
# Count missing values per column
missing_counts = df.isnull().sum()

# Calculate percentage of missing values
missing_percent = (missing_counts / len(df)) * 100

# Combine into a DataFrame for clarity
missing_report = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Percentage': missing_percent
}).sort_values(by='Missing Percentage', ascending=False)

# Display only columns with missing values
missing_report = missing_report[missing_report['Missing Count'] > 0]
display(missing_report)
```

```python
plt.figure(figsize=(15, 6))
sns.heatmap(df[missing_report.index].isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap')
plt.show()
```

```python
missing_report.to_csv("missing_value_report.csv", index=True)
```

```python
# Find exact duplicates
duplicate_rows = df[df.duplicated()]
print(f"Number of exact duplicate rows: {duplicate_rows.shape[0]}")

# Optionally display first few duplicates
display(duplicate_rows.head())
```

```python
# List of important numeric columns
numeric_cols = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
                'adults', 'children', 'babies', 'adr']

# Summary statistics
display(df[numeric_cols].describe())
```

```python
# Example: check unique values for a few categorical fields
cat_cols = ['meal', 'distribution_channel', 'market_segment', 'customer_type', 'deposit_type']

for col in cat_cols:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
```

```python
# Find rows where total guests = 0
df['total_guests'] = df['adults'] + df['children'].fillna(0) + df['babies']
invalid_guests = df[df['total_guests'] == 0]
print(f"Rows with zero total guests: {invalid_guests.shape[0]}")
display(invalid_guests.head())
```

```python
# Replace missing children values with 0
df['children'].fillna(0, inplace=True)

# Confirm it's clean
print("Missing in 'children':", df['children'].isnull().sum())
```

```python
# Check most frequent country
most_common_country = df['country'].mode()[0]
print("Most common country:", most_common_country)

# Option 1: Fill with mode
df['country'].fillna(most_common_country, inplace=True)

# Option 2 (alternative): Fill with 'Unknown'
# df['country'].fillna('Unknown', inplace=True)

# Confirm it's clean
print("Missing in 'country':", df['country'].isnull().sum())
```

```python
# Replace missing agent with 0, and convert to int
df['agent'].fillna(0, inplace=True)
df['agent'] = df['agent'].astype(int)

# Confirm
print("Missing in 'agent':", df['agent'].isnull().sum())
```

```python
# Replace missing company with 0, and convert to int
df['company'].fillna(0, inplace=True)
df['company'] = df['company'].astype(int)

# Confirm
print("Missing in 'company':", df['company'].isnull().sum())
```

```python
# Final check: Any remaining missing values?
missing_after = df.isnull().sum()
missing_after = missing_after[missing_after > 0]
print("Remaining missing values:\n", missing_after)
```

```python
df.to_csv("hotel_bookings_missing_cleaned.csv", index=False)
```

```python
# Function to find outliers using IQR
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return outliers

# Example: Outliers in 'adr'
adr_outliers = find_outliers_iqr(df, 'adr')
print(f"Outliers in 'adr': {adr_outliers.shape[0]}")
```

```python
# Remove 'adr' outliers above 500 (business domain: too high for daily rate)
df = df[df['adr'] < 500]

# Remove 'lead_time' outliers above 500 days
df = df[df['lead_time'] < 500]

print("New shape after outlier removal:", df.shape)
```
